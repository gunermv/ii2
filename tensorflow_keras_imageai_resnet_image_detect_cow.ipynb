```
import os
my_path='/content/MyDrive/MyDrive/python/'
from imageai.Detection import ObjectDetection
detector = ObjectDetection()
detector.setModelTypeAsRetinaNet()
detector.setModelPath(os.path.join(my_path, "resnet50_coco_best_v2.0.1.h5"))
detector.loadModel(detection_speed="fastest") # faster fast normal

```

```
custom = detector.CustomObjects(cow=True)
list = detector.detectCustomObjectsFromImage(
        custom_objects=custom,
        input_image=os.path.join(my_path, "objects.jpg"),
        output_image_path=os.path.join(my_path, "new_objects.jpg"),
        display_percentage_probability=False,
        display_object_name=False
    )

```
